```python
df = pd.read_csv('Auto.csv', na_values='?')
df.dropna(inplace=True)
df['horsepower'] = pd.to_numeric(df['horsepower'])
df.head()
```

name and origin are qualitative, the rest are quantitative

```python
#9.b
range_format = '{col} has a range of {rng}.'
for col in df.columns[:-2]:
    rng = df[col].max() - df[col].min()
    print(range_format.format(col=col, rng=rng))
```

```python
# 9.c
mean_std_format = '{col} has a mean of {mean} and a std of {std}.'
for col in df.columns[:-2]:
    mean = np.round(df[col].mean(), 2)
    std = np.round(df[col].std())
    print(mean_std_format.format(col=col, mean=mean, std=std))
```

```python
# 9.d
middle_idxs = df.index.isin(np.arange(10, 85))
limited_df = df[~middle_idxs]
```

```python
range_format = '{col} has a range of {rng}.'
for col in limited_df.columns[:-2]:
    rng = limited_df[col].max() - df[col].min()
    print(range_format.format(col=col, rng=rng))
```

```python
mean_std_format = '{col} has a mean of {mean} and a std of {std}.'
for col in limited_df.columns[:-2]:
    mean = np.round(limited_df[col].mean(), 2)
    std = np.round(limited_df[col].std())
    print(mean_std_format.format(col=col, mean=mean, std=std))
```
