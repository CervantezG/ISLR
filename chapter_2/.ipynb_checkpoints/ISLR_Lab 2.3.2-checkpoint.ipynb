```python
import numpy as np
import matplotlib.pyplot as plt 
```

Basic plotting

```python
np.random.seed(13)
x = np.random.normal(loc=0, size=100)
y = np.random.normal(loc=0, size=100)
plt.plot(x, y)
```

```python
plt.scatter(x, y)
```

```python
plt.savefig('scatter.png')
```

```python
plt.close()
```

Create some sequences

```python
# Note that the sequence does not include 10.
x = np.arange(1, 10)
x
```

```python
# Note that -pi and pi are both included.
x = np.linspace(-np.pi, np.pi, 50)
x
```

Create a contour plot

```python
y = x 
f = np.cos(y[:, None]) / (1 + x**2)
plt.contour(x, y, f)
```

```python
plt.contour(x, y, f, 45)
```

```python
fa = (f - np.transpose(f)) / 2
plt.contour(x, y, fa, 15)
```

Create a heatmap

```python
extent = np.min(x), np.max(x), np.min(y), np.max(y)
plt.imshow(fa, extent=extent)
```

Create surface plt

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

xx, yy = np.meshgrid(x, y)
surf = ax.plot_surface(xx, yy, fa)
```

```python
#Change angle
fig = plt.figure()
ax = fig.gca(projection='3d')

ax.view_init(30, 20)

xx, yy = np.meshgrid(x, y)
surf = ax.plot_surface(xx, yy, fa)
```
